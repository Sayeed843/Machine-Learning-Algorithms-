```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# darkgrid view graph
sns.set_style('darkgrid')
```

```python
size = [6, 8, 12, 14, 18]
price = [350, 775, 1150, 1395, 1675]
```

```python
pd.Series(price, size)
```

```python
df = pd.DataFrame({'size': size,'price': price})
```

```python
df
```

```python
fig = plt.figure(figsize=(8,8))
plt.scatter(x= df['size'], y=df['price'])

plt.xlabel("Priza Size (inc)")
plt.ylabel("Priza Price")
plt.title("A relation between pizza size and price")

# Annotate each point
for i, txt in enumerate(df['price']):
    plt.annotate(txt, (df['size'][i], df['price'][i]), textcoords="offset points",  xytext=(0,10), ha='center')

# More bins on x and y axes
plt.xticks(np.arange(0,21, 5))
plt.yticks(np.arange(0, 2000, 200))
    
plt.tight_layout()
plt.show()
```

```python
# we need to find out best line by using 
# -->  y=mx+c
```

```python
x=df['size'].sum()
y=df['price'].sum()
```

```python
xy=df['size']*df['price']
x2=df['size']*df['size']
```

```python
x_average=x/len(df['size'])
y_average=y/len(df['price'])
xy_average=xy.sum()/len(xy)
x2_average=x2.sum()/len(x2)
```

```python
'''
Now find the m value
Equcation- 
m = ((x_average* y_average) - xy_average)/(x_average*x_average - x2_average)
c = y_average - (m*x_average)
'''
```

```python
m = ((x_average* y_average) - xy_average)/(x_average*x_average - x2_average)
```

```python
m
```

```python
c = y_average - (m*x_average)
```

```python
c
```

```python
'''
So our line is,
y = mx+c
y = 107.21x - 107.21

This is our best line(hypothesis)
'''
```

```python
'''
If we want to find the price of 17inc priza, then we will use x=17
So, y is
y = 107.21m + c
y = 107.21*17 - 174.69
'''
```

```python
y = 107.21*17 - 174.69
```
